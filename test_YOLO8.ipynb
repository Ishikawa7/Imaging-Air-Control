```python
import cv2
import matplotlib.pyplot as plt
from IPython.display import clear_output
import plotly.express as px
import time
```

```python
import torch
from ultralytics import YOLO
```

```python
# load model
model = YOLO("yolov8n.pt")
```

```python
def display_cv_image(image):
    results = model(image)
    # display the image with cv2
    cv2.imshow('image', results[0].plot())
```

```python
def test_webcam_in_notebook():
    # Open the default camera (usually the first webcam)
    cap = cv2.VideoCapture(0)

    if not cap.isOpened():
        print("Error: Unable to open the webcam.")
        return

    try:
        while True:
            #clear_output(wait=True)
            # Capture frame-by-frame
            ret, frame = cap.read()
            display_cv_image(frame)
            # press q to exit
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
    finally:
        # Release the camera
        cap.release()
```

```python
test_webcam_in_notebook()
```

```python
# Images
imgs = ['image.jpg']  # batch of images

# Inference
results = model(imgs)  # includes NMS
```

```python
# Results
plt.imshow(cv2.cvtColor(results[0].plot(), cv2.COLOR_BGR2RGB))
```
